```python
import pandas as pd
excel=pd.read_excel("C:/Users/HA/Desktop/financial_data.xlsx")
def excel_to_csv(excel_file_path, csv_file_path):
    # Read the Excel file
    excel_data = pd.read_excel(excel_file_path)
    
    # Save as CSV file
    excel_data.to_csv(csv_file_path, index=False)  # index=False to exclude row numbers from the CSV

    print(f"Successfully converted {excel_file_path} to {csv_file_path}")

# Example usage
excel_to_csv("C:/Users/HA/Desktop/financial_data.xlsx", "C:/Users/HA/Desktop/financial_data.csv")
```
