```python
import pandas as pd
excel=pd.read_excel("C:/Users/HA/Desktop/financial_data.xlsx")
def excel_to_csv(excel_file_path, csv_file_path):
    # Read the Excel file
    excel_data = pd.read_excel(excel_file_path)
    
    # Save as CSV file
    excel_data.to_csv(csv_file_path, index=False)  # index=False to exclude row numbers from the CSV

    print(f"Successfully converted {excel_file_path} to {csv_file_path}")

# Example usage
excel_to_csv("C:/Users/HA/Desktop/financial_data.xlsx", "C:/Users/HA/Desktop/financial_data.csv")
```

```python
df=pd.read_csv("C:/Users/HA/Desktop/financial_data.csv")
df.head(9)
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenues(in millions)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income(in millions)'].pct_change() * 100
```

```python
average_revenue_growth = df.groupby('Company')['Revenue Growth (%)'].mean()
```

```python
max_net_income_growth = df.groupby('Company')['Net Income Growth (%)'].max()
min_net_income_growth = df.groupby('Company')['Net Income Growth (%)'].min()

print("Max Net Income Growth by Company:")
print(max_net_income_growth)
print("\nMin Net Income Growth by Company:")
print(min_net_income_growth)
```

```python

df['Assets Growth (%)'] = df.groupby('Company')['Total Assets(in millions)'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['total liabilities(in millions)'].pct_change() * 100
```

```python
df['Assets-to-Liabilities Ratio'] = df['Total Assets(in millions)'] / df['total liabilities(in millions)']
df['Cash Flow Growth (%)'] = df.groupby('Company')['cash flow from operating activities(in millions)'].pct_change() * 100
```

```python
df_summary = df[['Year', 'Company', 'Total Revenues(in millions)', 'Revenue Growth (%)', 'Net Income(in millions)', 'Net Income Growth (%)', 'Assets-to-Liabilities Ratio', 'Cash Flow Growth (%)']]
print(df_summary)
```

Apple experienced a slight decline in revenue growth from 2022 to 2023, with -2.29%, indicating potential stagnation in the company's growth trajectory.
2. Net Income Growth (2022-2023)
Tesla had outstanding net income growth in 2022, with a significant 127.5% increase, indicating very strong profitability. In 2023, the net income growth slowed but remained strong at 19.26%.
Microsoft had a positive net income growth of 18.7% in 2022, but the growth turned negative at -0.51% in 2023, showing stagnation in profitability.
Apple witnessed a huge 954% net income growth in 2022, though it slowed to -2.81% in 2023, suggesting some volatility in earnings.
3. Assets Growth
Tesla's assets saw a major jump in 2022 by 32.52%, but then in 2023, assets dropped significantly by -99.87%, possibly due to divestitures or a major restructuring.
Microsoft had consistent asset growth: 9.3% in 2022 and 12.91% in 2023, reflecting a steady expansion of the company's asset base.
Apple showed flat growth with 0.49% in 2022 and -0.04% in 2023, indicating that Apple may not have invested heavily in expanding its asset base.
4. Liabilities Growth
Tesla had an unusual drop in liabilities in 2022 (-88.07%), followed by an extreme increase in 2023 (1080%), which may indicate debt repayment followed by new debt issuance.
Microsoft showed relatively stable liabilities growth in 2022 and 2023 (around 3-4%).
Apple had a slight increase in liabilities in 2022 (4.92%), followed by a slight decrease in 2023 (-3.85%), indicating relatively stable management of debt.
5. Cash Flow from Operating Activities Growth
Tesla had a strong cash flow growth in 2022 (28%), followed by a decline of -9.97% in 2023, indicating possible changes in operational efficiency or investment strategy.
Microsoft's cash flow grew by 16% in 2022 but decreased by -1.63% in 2023, showing some fluctuation in the company's ability to generate cash from its core business.
Apple had positive cash flow growth in 2022 (17.4%), but this also decreased in 2023 by -9.5%.
6. Assets-to-Liabilities Ratio
Tesla's assets-to-liabilities ratio was extremely high in 2022 (22.59), reflecting very low liabilities compared to assets, but dropped significantly in 2023 to 0.0024, which indicates significant changes in the company’s balance sheet.
Microsoft and Apple maintained a stable assets-to-liabilities ratio between 1.16 to 1.84, showing that both companies have more assets than liabilities, which reflects a solid financial standing.
 SO ,Tesla showed extreme volatility in both assets and liabilities, possibly due to debt restructuring or aggressive expansion strategies, but continued to demonstrate strong growth.
Microsoft showed consistent growth in revenues and assets, though its net income growth and cash flow growth slowed in 2023.
Apple had a mixed performance, with declining revenue, net income, and cash flow growth in 2023, but maintained a stable balance sheet with relatively small changes in assets and liabilities.
